```python
#6)Apply spelling correction on each word tokens and print the initial 15 misspelled tokens as well as the corrected tokens. Keep a count of crrected tokens.
#from nltk import sent_tokenize

from autocorrect import Speller
filename="data.txt"
with open(filename, 'r', encoding="utf-8") as f:
    text=f.read()
    x= word_tokenize(text)
    
Spell=Speller(lang="en")
count=0
ListA=[]
for r in x:
    c=Spell(r)
    if r!=c and c not in ListA:
        count+=1
        print(str(count) +". "+r + ":"+c)
        ListA.append(c)
        if len(ListA)==15:
            break
 
```

```python
#Print All corrected tokens. Keep a count of corrected tokens
from autocorrect import Speller
filename="data.txt"
with open(filename, 'r', encoding="utf-8") as f:
    text=f.read()
    
    sentences = word_tokenize(text)
    words = [word.lower() for word in sentences if re.match('^[a-zA-Z]+', word)]  
    sptext=words
    
Spell=Speller(lang="en")
count=0
ListA=[]
for r in sptext:
    c=Spell(r)
    count+=1
    #print(c)
    ListA.append(c)
    
print(count) 
print(ListA)
      
```
